```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(n):
    if n == 0:
        return 200
    return round(f(n-1) + 0.5 * (1 - f(n-1)/2000)*f(n-1),0)
```

```python
y = []
x = []
for n in range(0,15):
    y.append(f(n))
    x.append(n)
```

```python
plt.plot(x,y)
plt.plot(x,y,marker='o')
plt.title("Population Vs Years")
plt.xticks(np.arange(min(x),max(x)+1,2))
plt.ylabel('Population')
plt.ylabel('Years')
```
